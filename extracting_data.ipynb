```python
import fitz  # PyMuPDF
import os

# Open the PDF
pdf = fitz.open('/content/TheoryOfComputation.pdf')

# Define the output folder
output_folder = '/content/images'
os.makedirs(output_folder, exist_ok=True)

# Process each page
for page_number in range(len(pdf)):
    page = pdf[page_number]
    images_on_page = page.get_images(full=True)
    print(f"Page {page_number + 1} contains {len(images_on_page)} images.")

    if len(images_on_page) == 0:
        print(f"No images found on page {page_number + 1}.")
        continue

    for image_index, image in enumerate(images_on_page):
        img_reference = image[0]
        try:
            base_image = pdf.extract_image(img_reference)
            image_bytes = base_image["image"]
            image_ext = base_image["ext"]
            image_filename = f"page_{page_number + 1}_image_{image_index + 1}.{image_ext}"
            image_path = os.path.join(output_folder, image_filename)

            # Write image file
            with open(image_path, "wb") as image_file:
                image_file.write(image_bytes)

            print(f"Image saved: {image_path}")
        except Exception as e:
            print(f"Failed to extract image on page {page_number + 1}, index {image_index}: {e}")

pdf.close()
print("Processing completed.")

```

```python
import fitz  # PyMuPDF
import os

# Path to your PDF file
pdf_path = '/content/TheoryOfComputation.pdf'

# Open the PDF
pdf = fitz.open(pdf_path)

# Define the output folders for images and vector content
image_folder = '/content/images/'
vector_folder = '/content/vector_graphics/'

# Create the folders if they don't exist (check if it's a directory)
os.makedirs(image_folder, exist_ok=True)
os.makedirs(vector_folder, exist_ok=True)

# Iterate through all pages in the PDF
for page_number in range(len(pdf)):
    page = pdf.load_page(page_number)  # Load the page

    # Extract raster images (JPEG, PNG)
    images_on_page = page.get_images(full=True)

    for image_index, image in enumerate(images_on_page):
        img_reference = image[0]  # Get the xref number
        base_image = pdf.extract_image(img_reference)  # Extract image data
        image_bytes = base_image["image"]  # Raw image bytes
        image_ext = base_image["ext"]  # Image file extension (PNG, JPEG, etc.)

        # Save the raster image to a file
        image_filename = f"page_{page_number + 1}_image_{image_index + 1}.{image_ext}"
        image_path = os.path.join(image_folder, image_filename)
        with open(image_path, "wb") as image_file:
            image_file.write(image_bytes)
        print(f"Image saved: {image_path}")

    # Extract vector graphics (SVG format)
    vector_text = page.get_text("svg")  # Extract vector content as SVG (can also use "html" or "text")
    if vector_text:
        vector_filename = f"page_{page_number + 1}_vector.svg"
        vector_path = os.path.join(vector_folder, vector_filename)

        #
        # Save vector content as SVG
        with open(vector_path, 'w') as vector_file:
            vector_file.write(vector_text)
        print(f"Vector content saved: {vector_path}")

# Close the PDF file
pdf.close()

print("Image and vector extraction completed.")
```

```python
import fitz  # PyMuPDF
import os

# Path to your PDF file
pdf_path = '/content/TheoryOfComputation.pdf'

# Open the PDF
pdf = fitz.open(pdf_path)

# Define the output folders for images and vector content
image_folder = '/content/images/'
vector_folder = '/content/vector_graphics/'

# Create the folders if they don't exist
os.makedirs(image_folder, exist_ok=True)
os.makedirs(vector_folder, exist_ok=True)

# Iterate through all pages in the PDF
for page_number in range(len(pdf)):
    page = pdf.load_page(page_number)  # Load the page

    # Extract raster images (JPEG, PNG)
    images_on_page = page.get_images(full=True)

    if images_on_page:  # If there are raster images on the page
        for image_index, image in enumerate(images_on_page):
            img_reference = image[0]  # Get the xref number
            base_image = pdf.extract_image(img_reference)  # Extract image data
            image_bytes = base_image["image"]  # Raw image bytes
            image_ext = base_image["ext"]  # Image file extension (PNG, JPEG, etc.)

            # Save the raster image to a file
            image_filename = f"page_{page_number + 1}_image_{image_index + 1}.{image_ext}"
            image_path = os.path.join(image_folder, image_filename)
            with open(image_path, "wb") as image_file:
                image_file.write(image_bytes)
            print(f"Image saved: {image_path}")
    else:
        # If no images, consider saving the page as a full-page image (rasterized)
        pix = page.get_pixmap()  # Render page to a pixmap (image)
        image_filename = f"page_{page_number + 1}_full_page.png"  # Save as PNG
        image_path = os.path.join(image_folder, image_filename)
        pix.save(image_path)
        print(f"Full page image saved: {image_path}")

    # Extract vector graphics (SVG format)
    vector_text = page.get_text("svg")  # Extract vector content as SVG (can also use "html" or "text")
    if vector_text:
        vector_filename = f"page_{page_number + 1}_vector.svg"
        vector_path = os.path.join(vector_folder, vector_filename)

        # Save vector content as SVG
        with open(vector_path, 'w') as vector_file:
            vector_file.write(vector_text)
        print(f"Vector content saved: {vector_path}")

# Close the PDF file
pdf.close()

print("Image and vector extraction completed.")
```

```python
from paddleocr import PaddleOCR
import cv2
import numpy as np

# 1. OCR Modelini Yükle
ocr = PaddleOCR()

# 2. Görüntüyü Yükle
img_path = '/content/images/page_44_full_page.png'
img = cv2.imread(img_path)

# 3. Metin Algılama
results = ocr.ocr(img_path)

# 4. Metin Alanlarını Maskeleme
for line in results[0]:
    box = line[0]
    box = [(int(x), int(y)) for x, y in box]
    cv2.fillPoly(img, [np.array(box, dtype=np.int32)], (0, 0, 0))

# 5. Sonucu Kaydetme
cv2.imwrite('processed_page.png', img)
```

```python
from PIL import Image
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
# Load the uploaded image
input_path = "/content/processed_page.png"
image = cv2.imread(input_path)

# Convert image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply thresholding to isolate shapes
_, thresh = cv2.threshold(gray, 200, 255, cv2.THRESH_BINARY_INV)

# Find contours
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Sort contours based on their area and keep the largest three (assume these are the shapes)
sorted_contours = sorted(contours, key=cv2.contourArea, reverse=True)[:3]

# Create a directory to save the cropped shapes
output_paths = []

if(os.path.exists('/content/shapes')==False):
  os.mkdir('/content/shapes')
for i, contour in enumerate(sorted_contours):
    # Get bounding box for each contour
    x, y, w, h = cv2.boundingRect(contour)
    cropped_shape = image[y:y+h, x:x+w]

    # Save each shape as a separate file
    output_path = f"/content/shapes/shape_{i+1}.png"
    cv2.imwrite(output_path, cropped_shape)
    output_paths.append(output_path)

# Show the output paths
output_paths
```
